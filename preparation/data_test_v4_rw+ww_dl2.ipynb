```python
# Separate target variable Y from features X
print("Separating labels from features...")
target_variable = "quality"
# X = wines.ix[:, 0:11]
# y = np.ravel(wines.type)
X = wines.drop(target_variable, axis = 1)
Y = wines.loc[:,target_variable]

print("...Done.")
print()

print('Y : ')
print(Y.head())
print()
print('X :')
X.head()
```

```python
# Divide dataset Train set & Test set 
print("Dividing into train and test sets...")
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, stratify=Y)
print("...Done.")
print()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers
from keras.layers import Dropout

model = Sequential()
model.add(Dense(64, input_dim=12, activation='relu'))
# model.add(Dropout(0.05))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(64, activation='relu'))
# model.add(Dropout(0.05))
model.add(Dense(32, activation='relu'))
# model.add(Dropout(0.03))
model.add(Dense(32, activation='relu'))
# model.add(Dropout(0.05))
model.add(Dense(16, activation='relu'))
# model.add(Dropout(0.05))
model.add(Dense(16, activation='relu'))
# model.add(Dropout(0.05))
model.add(Dense(8, activation='relu'))
# model.add(Dense(8, activation='relu'))
model.add(Dropout(0.05))
model.add(Dense(1, activation='linear'))
#  kernel_regularizer=regularizers.l1_l2(0.01)
# Compiler le modèle
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])
# 'mean_squared_error'
# Créer un modèle de deep learning simple
# model = Sequential()
# model.add(Dense(64, activation='relu', input_dim=12))
# model.add(Dense(64, activation='relu'))
# model.add(Dense(12, activation='softmax'))

# Compiler le modèle
# model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Définir l'ensemble de validation
validation_data = (X_val, Y_val)

# Entraîner le modèle
history = model.fit(X_train, Y_train, epochs=150, batch_size=16, validation_data=validation_data)

# Tracer la perte par epoch
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='validation')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
